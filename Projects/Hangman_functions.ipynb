```python
import random


def hangman_game():

    gallows = ['''
     +---+
      |   |
          |
          |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
          |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
      |   |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|   |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\  |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\  |
     /    |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\  |
     / \  |
          |
    =========''']

    def correct_guess():
        if guess in chosen:
            return True
        else:
            return False

    def get_gallows():
        return gallows[guess_count]

    word_list = ['ant', 'frog', 'boom', 'chair', 'brush', 'light', 'donkey', 'jumper', 'quacks', 'zephyr',
                 'picture', 'country', 'junkmail', 'brought', 'showjumping']

    def get_word():
        global selected
        selected = random.choice(word_list)
        return [char for char in selected]

    chosen = get_word()
    char_lines = []
    for char in chosen:
        char_lines.append(" _ ")
    guess_count = 0
    guess_limit = len(gallows)
    wrong_guess = []

    print(get_gallows())
    print(wrong_guess)
    print(char_lines)

    while guess_count <= guess_limit:
        guess = input("Enter Guess: ")
        if correct_guess() == True:
            print(get_gallows())
            print(wrong_guess)
            guess_pos = [i for i in range(len(chosen)) if chosen[i] == guess]
            for j in guess_pos:
                char_lines.pop(j)
                char_lines.insert(j, guess)
            print(char_lines)
        if char_lines == chosen:
            print("You did it")
            print(f"The word was {selected}")
            break
        elif correct_guess() == False:
            guess_count += 1
            if guess_count < guess_limit:
                if guess in wrong_guess:
                    print("you've already guessed this")
                else:
                    wrong_guess.append(guess)
                print(get_gallows())
                print(wrong_guess)
                print(char_lines)
            else:
                print("Try again")
                print(f"The word was {selected}")
                break




hangman_game()
```
