```python
# variables with gallows at all stages
# word to guess with spaces
# list of incorrect guess
import random


def hang_man_lines():
    HANGMANPICS = ['''
  +---+
  |   |
      |
      |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========''']

    word_list = ['ant', 'frog', 'boom', 'chair', 'brush', 'light', 'donkey', 'jumper', 'quacks', 'zephyr'
                 'picture', 'country', 'junkmail', 'brought', 'showjumping']
    secret_word = random.choice(word_list)
    guess_count = 0
    guess_limit = len(HANGMANPICS)
    wrong_guess = []

    def split(secret_word):
        return [char for char in secret_word]

    to_guess = split(secret_word)
    # print(to_guess)
    char_lines = []
    for char in to_guess:
        char_lines.append(" _ ")
    # print(char_lines)
    print(HANGMANPICS[guess_count])
    print(wrong_guess)
    print(char_lines)
    while guess_count <= guess_limit:
        guess = input("Guess a letter: ")
        if guess in to_guess:
            print(HANGMANPICS[guess_count])
            print(wrong_guess)         
            guess_pos = [ i for i in range(len(to_guess)) if to_guess[i] == guess]
            for y in guess_pos:
                char_lines.pop(y)
                char_lines.insert(y, guess)
            print(char_lines)
        if guess_count > guess_limit:
            print("Try again")
            print(f"The word was {secret_word}")
            break
        if char_lines == to_guess:
            print("You did it")
            print(f"The word was {secret_word}")
            break
        elif guess not in to_guess:
            guess_count += 1
            if guess_count < guess_limit:
                wrong_guess.append(guess)
                print(HANGMANPICS[guess_count])
                print(wrong_guess)
                print(char_lines)
            else:
                print("Try again")
                print(f"The word was {secret_word}")
                break


hang_man_lines()
```
